```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data_sources/cannabis.csv')
```

```python
df.head(20)
```

```python
df['Effects'] = df['Effects'].apply(lambda k: np.NaN if k == 'None' else k)
```

```python
df['Flavor'] = df['Flavor'].apply(lambda k: np.NaN if k == 'None' else k)
```

```python
df['Strain'] = df['Strain'].str.replace('-', ' ')
```

```python
df = df.dropna().reset_index().drop('index', axis=1)
```

What's the Goal?
- Target -> Strain
- X: Effects & Flavor
- Description -> Score?

```python
for column_name in ('Effects', 'Flavor'):
    df[column_name] = df[column_name].str.split(',')
```

```python
""" Not used but interesting """
def explode_dummies(df: pd.DataFrame, column_name: str) -> pd.DataFrame:
    if type(df[column_name][0]) == str:
        df[column_name] = df[column_name].str.split(',')        
    temp = pd.get_dummies(df[column_name].explode()).sum(level=0)
    new_df = pd.concat([df, temp], axis=1)
    return new_df
```
