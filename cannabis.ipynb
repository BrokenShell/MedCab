```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cannabis.csv")
```

```python
df.head(20)
```

```python
df['Rating'] = df['Rating'].apply(lambda s: np.NaN if s == 0.0 else s)
```

```python
df['Effects'] = df['Effects'].apply(lambda s: np.NaN if s == "None" else s)
```

```python
df['Flavor'] = df['Flavor'].apply(lambda s: np.NaN if s == "None" else s)
```

```python
df = df.dropna()
df.head(10)
```

```python
raw_score = SentimentIntensityAnalyzer()

def sentiment_score(sentence: str) -> int:
    sentiment_dict = raw_score.polarity_scores(sentence)
    return (100 + sentiment_dict['compound'] * 101) // 2
```

```python
df['Sentiment'] = df['Description'].apply(sentiment_score).astype(int)
```

```python
print(f"Sentiment Score Range: {df['Sentiment'].min()} - {df['Sentiment'].max()}")
```

```python
df.isna().sum()
```

```python
df = df.sort_values(by=['Rating', 'Sentiment'], ascending=False)
df = df.reset_index().drop(columns=['index'])
df.head(10)
```

```python
df[df['Strain'].isin([
    'Sour-Diesel', 
    'Wedding-Cake', 
    'God-Bud',
])].sort_values(by='Sentiment', ascending=False)
```
